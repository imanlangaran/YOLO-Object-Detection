```python
from imageai.Detection import ObjectDetection, VideoObjectDetection

detector = ObjectDetection()
```

```python
detector.setModelTypeAsYOLOv3()
```

```python
filepath = 'https://github.com/OlafenwaMoses/ImageAI/releases/download/3.0.0-pretrained/yolov3.pt/'
filename = 'yolov3.pt'
```

```python
import os
import urllib.request
```

```python
urllib.request.urlretrieve(filepath,filename)
```

```python
detector.setModelPath(filename)
```

```python
detector.loadModel()
```

```python
detection2 = detector.detectObjectsFromImage(input_image='image1.jpg', output_image_path="image2_detect.jpg", minimum_percentage_probability=80)
detection2
```

```python
custom = detector.CustomObjects(handbag=True, car=True)
detection3 = detector.detectObjectsFromImage(custom_objects=custom, input_image='image1.jpg', output_image_path="image3_detect.jpg", minimum_percentage_probability=80)
detection3
```
