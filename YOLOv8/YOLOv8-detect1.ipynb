```python
from ultralytics import YOLO
from PIL import Image
from IPython.display import Image as img
```

```python
model = YOLO('yolov8n.pt')
```

```python
results = model.predict(source='/content/drive/MyDrive/Colab Notebooks/datasets/bus.jpg',save=True, save_txt=True, save_conf=True, conf=0.5)
```

```python
im_arr = results[0].plot()
```

```python
im = Image.fromarray(im_arr[...,::-1])
```

```python
im.save('res1.jpg')
```

```python
im.show()
```
